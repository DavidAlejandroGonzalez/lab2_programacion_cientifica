```python
#importar las librerias
from functools import reduce
import random
```

```python
# Definición de la función lambda
expresion = lambda x: x**2 + 2*x + 1 

# Pruebas con valores 2, 4 y 6
valores_prueba = [2, 4, 6] 
resultados = [expresion(v) for v in valores_prueba]

print(f"Resultados para 2, 4 y 6: {resultados}")
```

```python
velocidades = [10, 20, 30, 40, 50] 

# Conversión: km/h -> m/s
velocidades_ms = list(map(lambda v: v / 3.6, velocidades))

print(f"Velocidades en m/s: {velocidades_ms}")
```

```python
valores = [2, 5, 8, 11, 14, 17, 20] 

# Filtrar pares (aquellos cuyo resto al dividir por 2 es 0)
pares = list(filter(lambda x: x % 2 == 0, valores)) 

print(f"Números pares: {pares}")
```

```python
numeros_prod = [2, 3, 4, 5]

# Multiplicación acumulada
producto_total = reduce(lambda x, y: x * y, numeros_prod)

print(f"Producto total: {producto_total}")
```

```python
data = [12, 15, 22, 35, 41, 8, 3, 55, 62, 90] 

# Pipeline: Filtrar < 10 -> Multiplicar x 2 -> Sumar todo
resultado_pipeline = reduce(lambda x, y: x + y, 
                     map(lambda x: x * 2, 
                     filter(lambda x: x >= 10, data))) 

print(f"Resultado del pipeline: {resultado_pipeline}")
```

```python
# Generación de datos
big_data = [random.randint(0, 100) for _ in range(1000000)] 

# 1. Filtrar > 50
filtrados = list(filter(lambda x: x > 50, big_data)) 

# 2. Elevar al cuadrado
cuadrados = list(map(lambda x: x**2, filtrados)) 

# 3. Promedio final
promedio = sum(cuadrados) / len(cuadrados) if cuadrados else 0 

print(f"Promedio del dataset procesado: {promedio}")
```

```python
# 500,000 lecturas de sensores (10°C a 80°C)
sensores = [random.uniform(10, 80) for _ in range(500000)] 

# Procesamiento en pipeline:
# Filtro rango 10-70 
datos_limpios = list(filter(lambda t: 10 <= t <= 70, sensores)) 
#convertir a kelvin
datos_kelvin = list(map(lambda t: t + 273.15, datos_limpios)) 
#promedio
promedio_k = sum(datos_kelvin) / len(datos_kelvin) if datos_kelvin else 0 

print(f"Temperatura promedio en Kelvin: {promedio_k:.2f} K")
```
